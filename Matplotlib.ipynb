```python
# Sine Curve

x = np.arange(0 , 4*np.pi , 0.2)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize = (12,6))
plt.title('sinx vs cosx') 

plt.subplot(2,1,1)
plt.ylabel('sinx')
plt.xlabel('Radians')
plt.plot(x, y_sin  , linestyle = 'dashed' , color = 'red')

plt.subplot(2,1,2)
plt.ylabel('cosx')
plt.xlabel('Radians')


plt.plot(x, y_cos , linestyle = 'dashed' , color = 'blue')        
plt.show()
```
